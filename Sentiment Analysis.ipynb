```python
auth = tweepy.OAuthHandler(api_key, api_secret_key)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
result = api.search(q="PPKM DKI Jakarta", lang ='id', count=100)
```

```python
HasilAnalisis = []

for tweet in result:
    prop = {}
    prop['Tanggal_Tweet'] = tweet.created_at
    prop['User'] = tweet.user.screen_name
    prop['Content'] = tweet.text
    cleaned = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet.text).split())
    
    analysis = TextBlob(cleaned)
    try:
        analysis = analysis.translate(to='en')
    except Exception as e:
        print(e)
    
    if analysis.sentiment.polarity > 0.0:
        prop['Sentiment'] = 'Positive'
    elif analysis.sentiment.polarity == 0.0:
        prop['Sentiment'] = 'Neutral'
    else :
        prop['Sentiment'] = 'Negative'
    
    if tweet.retweet_count > 0:
        if prop not in HasilAnalisis:
            HasilAnalisis.append(prop)
    else:
        HasilAnalisis.append(prop)
```

```python
pos = [t for t in HasilAnalisis if t['Sentiment']=='Positive']
net = [t for t in HasilAnalisis if t['Sentiment']=='Neutral']
neg = [t for t in HasilAnalisis if t['Sentiment']=='Negative']
```

```python
print('Sentiment Result')
print('Positive: ',len(pos), '({}%)'.format(len(pos)/len(HasilAnalisis)*100))
print('Neutral: ',len(net), '({}%)'.format(len(net)/len(HasilAnalisis)*100))
print('Negative: ',len(neg), '({}%)'.format(len(neg)/len(HasilAnalisis)*100))
```
